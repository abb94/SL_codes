```python
import os
import cv2
import matplotlib.pyplot as plt
from natsort import natsorted

# Directory containing frames
frame_dir = f"C:/Users/Bruss/Desktop/Speciale/data/data_for_pic"


# Get list of frame file names, sorted numerically
frame_names = natsorted(os.listdir(frame_dir), key=lambda x: int(x.split(".")[0]))

# Plot every 10th image
fig, ax = plt.subplots(1, len(frame_names) // 3 + 1, figsize=(20, 5), sharey=True)
for i, frame_name in enumerate(frame_names):
    if i % 3 == 0:
        # Read frame
        frame_path = os.path.join(frame_dir, frame_name)
        frame = cv2.imread(frame_path)

        # Convert BGR to RGB
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Plot frame
        ax[i // 3].imshow(frame)
        ax[i // 3].set_title(frame_name)
        ax[i // 3].axis("off")

# Save plot to file
output_path = "/path/to/output.png"
plt.savefig(output_path)

# Show plot
plt.show()
```
